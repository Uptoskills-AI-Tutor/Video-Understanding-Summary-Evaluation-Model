```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Reference summary - ideal summary for your video
reference_summary = "This video explains the concept of similarity scoring to evaluate understanding."

def evaluate_summary(user_summary):
    vectorizer = TfidfVectorizer().fit([user_summary, reference_summary])
    vectors = vectorizer.transform([user_summary, reference_summary])
    similarity = cosine_similarity(vectors[0], vectors[1])[0][0]
    return similarity

# Input from user
user_summary = input("Write your one-line summary of the video: ")

score = evaluate_summary(user_summary)
print(f"\nSimilarity score: {score:.2f}")

if score < 0.5:
    print("You haven't properly understood the video.")
elif score == 0.5:
    print("Please re-watch the video.")
elif score > 0.75:
    print("Great! You have a good understanding of the content.")
else:
    print("Your understanding is average. Consider revisiting some parts.")
```
